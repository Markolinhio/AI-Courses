## 1/ Điều kiện if-else

Trong Python, các câu điều kiện if-else được sinh ra để phục vụ mục đích xử lý một phép toán khi phép toán đó thỏa một điều kiện nhất định.  

Ví dụ phân loại học sinh trong lớp: Nếu học sinh có điểm trung bình bằng hoặc trên 8 thì là học sinh giỏi, còn nếu điểm trung bình từ 6.5 đến 8 thì là học sinh khá, còn lại là học sinh trung bình. Lúc đấy, có thể viết câu điều kiện if-else như sau (lưu ý các dấu ```:``` trong cú pháp và cách căn lề):
```python
if diem_trung_binh >= 8.:
    print('Học sinh Giỏi')
elif 6.5 < diem_trung_binh < 8.:
    print('Học sinh Khá')
else:
    print('Học sinh Trung Bình')
```

Chúng ta có thể thấy thì trong cú pháp câu điều kiện if-else, sẽ bắt đầu bằng điều kiện ```if``` đi kèm với một điều kiện cụ thể (ở đây là điểm >= 8) và theo đó là một dấu ```:```. Nếu như có thêm nhiều điều kiện sau if thì ta dùng ```elif``` (else if) đi kèm với một điều kiện cụ thể (ở đây là điểm > 8 và điểm < 6.5) cùng một dấu ```:```. Cuối cùng là phần còn lại sẽ đặt sau dòng ```else``` và dấu ```:```.

Lưu ý, trong điều kiện if-else, các vòng lặp loop, và trong các hàm hay lớp, nội dung của các điều kiện, vòng lặp, v.v. này đều phải được đặt cách dòng vào như ví dụ trên. Cách dòng có thể dùng nút Tab hoặc nút space (chỉ được dùng một trong hai chứ không dùng cả 2 như nhau vì Python rất quan trọng chuyện thụt dòng). Thường thì một hoặc 2 nút Tab là đủ để code nhìn đẹp.

## 2/ Vòng lặp while loop

Vòng lặp while loop có thể hiểu như là: Trong khi một điều kiện còn đang đúng thì lặp lại một phép tính liên tục.  

Ví dụ: chia đôi một số n=100 t lần cho đến khi số $n_{t} < 1$. Lúc đấy, ta có thể viết vòng while loop như sau (lưu ý các dấu ```:``` và cách căn thụt lề):
```python
n = 100
t = 0
while n > 1: # xác định tiếp tục chia đôi cho đến khi số n < 1
    n = n / 2 # chia đôi số n
    t = t + 1 # mỗi lần tính n/2, tăng t lên 1 để tính số lần t
```

## 3/ Vòng lặp for loop

Trước khi đi vào for loop, ta có thể nghiên cứu qua một hàm rất quan trọng và thông dụng trong Python, đó là hàm range.  
Khi khai báo một hàm range, ta dùng cú pháp: ```range(i, j)```. Range sẽ cho chúng ta một ```range``` đi từ ```i``` đến ```j-1```.

Chúng ta dùng từ ```range``` vì bản chất hàm này không trả ra một list đi từ ```i``` đến ```j-1``` (không phải kiểu dữ liệt list mà là kiểu dữ liệu range). Để minh họa:
```python
range(1, 5) # python sẽ trả ra output như bên dưới. chúng ta sẽ không có một list mà chỉ là một kiểu dữ liệu range(1, 5) như hình
>>>>> range(1, 5)

list(range(1, 5)) # tuy nhiên khi quy đổi kiểu dữ liệu range thành list, ta sẽ được output là một list như bình thường
>>>>> [1, 2, 3, 4]
```

Khi dùng trong for loop, Python sẽ tự hiểu chúng ta cần lặp từ ```i``` đến ```j-1```.

Tuy nhiên, nếu chúng ta cần một list thì chúng ta có thể biến đổi kiểu dữ liệu từ range thành list như sau:
```
list(range(i, j))
```

Vòng lặp for loop có thể hiểu như là: trong một số lần đã quy định, lặp lại một phép tính liên tục.

Ví dụ: chia đôi số n = 100 5 lần. Lúc đấy ta có thể viết dòng for loop như sau (lưu ý các dấu ```:``` và cách căn thụt lề):

```python
n = 100
for i in range(0, 5): # xác định chạy 5 lần
    n = n / 2 # chia đôi số n
```

Ngoài ra, còn nhiều cách sử dụng for loop khác như:
Ví dụ 2: nhân đôi các phần tử trong một list:

```python
list_a = [1, 2, 3, 4, 5]
list_b = [] # khởi tạo list_b rỗng để chứa các phần tử đã nhân đôi trong list_a
for phan_tu in list_a: # lặp qua các phần tử trong list a
    list_b.append(phan_tu * 2) # nhân đôi phần tử và thêm vào list_b mới
```

Ví dụ 3: đi qua các key-value trong một dict, nếu key chia hết cho 2 và value có độ dài lớn hơn 3 thì in ra:

```python
dict_a = {1 : [1, 2, 3], 2 : [2, 1], 3 : [1, 4, 6, 9], 4 : [1, 2, 3, 4]}
# vì dict_a.items() cho ra một cặp key, value, ta có thể đi qua vòng lặp bằng cách khai báo một cặp giá trị tương ứng
for key, value in dict_a.items():
    if key % 2 == 0 and len(value) > 3: # điều kiện key chia hết cho 2 và độ dài value > 3
        print(key, value) # in ra cặp key value thỏa điều kiện
```
