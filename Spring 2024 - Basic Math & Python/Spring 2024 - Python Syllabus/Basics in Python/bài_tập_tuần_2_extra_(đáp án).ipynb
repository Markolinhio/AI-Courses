### Bài 1: Xúc xắc
Cho 2 xúc xắc, mỗi xúc xắc có 6 mặt và các giá trị của các mặt biểu diễn là một dạng list độ dài 6, lần lượt:
+ Red: [3,3,3,3,3,6]
+ Blue: [2,2,2,5,5,5]
+ Green: [1,4,4,4,4,4]

Luật chơi như sau: mỗi người chơi chọn 1 trong 3 loại xúc xắc (chọn bằng cách nhập 'r' cho RED, 'b' cho BLUE, và 'g' cho GREEN) và quy định một số n lần lắc xúc xắc nhất định. Người chơi chiến thắng là người có số điểm lắc được lớn hơn người kia. Nói cách khác, trong n lần lắc, người chơi chiến thắng là người chơi có số lần thắng nhiều hơn người còn lại.

Có 2 cách tính điểm, singles và doubles:
+ Singles: trong một lượt, mỗi người lắc một lần và thắng thua dựa trên giá trị của mặt xúc xắc
+ Doubles: trong một lượt, mỗi người lắc hai lần và thắng thua dựa trên tổng giá trị của 2 lần lắc

Bài tập: Hoàn thành hàm simulate_singles và simulate_doubles để tính điểm. Và hoàn thành hàm simulate_and_print_result để in ra kết quả của cả 2 người chơi

```python
RED = [3,3,3,3,3,6]
BLUE = [2,2,2,5,5,5]
GREEN = [1,4,4,4,4,4]
DICE_NAMES = ["Red", "Blue", "GREEN"]

# TODO: Hoàn thành phần code ở dưới
# die là tên xúc xắc người chơi chọn, quy định là 'r', 'b', hoặc 'g'
# number là mặt xúc xắc, chọn ngẫu nhiên bởi máy
# Hoàn thành hàm roll để lắc xúc xắc tương ứng và trả ra mặt của xúc xắc tương ứng (dùng if-else)
# Ví dụ: mặt số 1 của xúc xắc Red là RED[0] 
def roll(die):
    number = random.randint(0,5)
    if die == 'r':    
        return RED[number]
    elif die == 'b':
        return BLUE[number]
    else:
        return GREEN[number]
    
# TODO: Hoàn thành phần code ở dưới. 
# Trong số lần roll quy định sẵn, tính số ván thắng của 2 người chơi và số ván hòa
def simulate_singles(die1, die2, rolls):
    win1 = 0
    win2 = 0
    draw = 0
    # TODO: Lắc xúc xắc theo số lần đã cho và tính số ván thắng của 2 người chơi và số ván hòa
    for i in range (0, rolls):
        a = roll(die1)
        b = roll(die2)
        if (a > b):
            win1 += 1
        elif (a < b):
            win2 += 1
        else:
            draw += 1
    return win1, win2, draw

# TODO: Hoàn thành phần code ở dưới. 
# Trong số lần roll quy định sẵn, tính số ván thắng của 2 người chơi và số ván hòa
# Đây là thể doubles, nên là mỗi người chơi lắc hai lần và thắng thua dựa trên tổng 2 lần lắc
def simulate_doubles(die1, die2, rolls):
    win1 = 0
    win2 = 0
    draw = 0
    for i in range (0, rolls):
        a = roll(die1)
        b = roll(die1)
        c = roll(die2)
        d = roll(die2)
        if ((a+b) > (c+d)):
            win1 += 1
        elif ((a+b) < (c+d)):
            win2 += 1
        else:
            draw += 1
    return win1, win2, draw

# TODO: Hoàn thành phần code ở dưới, in ra mỗi người chơi thắng mấy lần với tỉ lệ bao nhiêu. 
# Đồng thời nếu hòa thì in ra tỉ lệ hòa là bao nhiêu
# Lưu ý: 2 parameter:
# simulation_function: hàm tính điểm (singles hoặc doubles)
# header: Biểu diễn đây là ván single hay doubles
def simulate_and_print_result(die1, die2, rolls, simulation_function, header):
    wins1, wins2, draws = simulation_function(die1, die2, rolls)
    print(header)
    print("Player 1 used {:s} die and won {:d} times, so {:.1f}% of the rolls.".format(DICE_NAMES[die1],wins1,wins1/rolls*100))
    print("Player 2 used {:s} die and won {:d} times, so {:.1f}% of the rolls.".format(DICE_NAMES[die2],wins2,wins2/rolls*100))
    if draws != 0:
        print("{:d} draws, so {:.2f}% of the rolls.".format(draws, draws/rolls*100))
```

```python
print("Welcome to a non-transitive dice simulation.")
print("The dice:")
print("{:d} for {:s}: {:}".format('r' ,DICE_NAMES[0], RED))
print("{:d} for {:s}: {:}".format('b' ,DICE_NAMES[1], BLUE))
print("{:d} for {:s}: {:}".format('g' ,DICE_NAMES[2], GREEN))

choice1 = str(input("Choose a die for player 1:\n"))
while choice1 not in ['r', 'b', 'g']:
    choice1 = str(input("Choose a die for player 1:\n"))
choice2 = int(input("Choose a die for player 2:\n"))
while choice2 not in ['r', 'b', 'g']:
    choice2 = str(input("Choose a die for player 2:\n"))
rolls  = int(input("How many rolls to simulate?\n"))
simulate_and_print_result(choice1, choice2, rolls, simulate_singles, "Singles:")
simulate_and_print_result(choice1, choice2, rolls, simulate_doubles, "Doubles:")
```

### Bài 2: So sánh từ
Cho 2 từ có độ dài từ 3 đến 7 ký tự.
Thực hiện so sánh 2 từ đó sao cho:
+ 2 từ có độ dài khác nhau, trả ra 2 list rỗng
+ 2 từ có độ dài giống nhau: trả ra 2 list sao cho:
  - List 1: Chứa vị trí của các chữ cái giống nhau và cùng 1 vị trí trên 2 từ
  - List 2: Chứa 2 list, mỗi list là vị trí của các chữ cái giống nhau và khác vị trí trên 2 từ

```python
# Generate two words with the specified conditions
def generate_two_words(words):
    # Fetch a list of English words from nltk
    def get_words(words):
        word_list = words.words()
        return [word.lower() for word in word_list if 3 <= len(word) <= 7]
    lower_cased_words = get_words(words)

    while True:
        word1 = random.choice(lower_cased_words)
        word2 = random.choice(lower_cased_words)
        
        # Ensure the two words have the same length 50% of the time
        if random.random() < 0.6:
            word2 = random.choice([w for w in lower_cased_words if len(w) == len(word1)])
        
        # Check for at least 2 common letters in the same position
        common_positions = sum(1 for i in range(min(len(word1), len(word2))) if word1[i] == word2[i])
        if common_positions >= 2:
            return word1, word2
        
word1, word2 = generate_two_words(words)
word1, word2
```

```python
# TODO: So sánh 2 từ với nhau sao cho:
# 2 từ có độ dài khác nhau, trả ra 2 list rỗng
# 2 từ có độ dài giống nhau: trả ra 2 list
# List 1 chứa vị trí của các chữ cái giống nhau và cùng 1 vị trí trên 2 từ
# List 2 chứa 2 list
# Mỗi list chứa vị trí của các chữ cái giống nhau 
# Nhưng không cùng vị trí trên 2 từ trên lần lượt từ 1 và từ 2
def compare_words(word1, word2):
    same_letters_same_place = []
    same_letters_different_place_word1 = []
    same_letters_different_place_word2 = []
    
    # TODO: trả ra các kết quả tương ứng
    if len(word1) != len(word2):
        return [], []
    else:
        word1 = word1.lower()
        word2 = word2.lower()
        
        # TODO: Trả ra List 1 chứa vị trí của các chữ cái giống nhau và cùng 1 vị trí trên 2 từ
        for i in range(len(word1)):
            if word1[i] == word2[i]:
                same_letters_same_place.append(i)
        
        # TODO: Trả ra List 2 chứa 2 list 
        # Mỗi list chứa vị trí của các chữ cái giống nhau 
        # Nhưng không cùng vị trí trên 2 từ trên lần lượt từ 1 và từ 2
        for i in range(len(word1)):
            for j in range(len(word2)):
                if i != j and word1[i] == word2[j]:
                    same_letters_different_place_word1.append(i)
                    same_letters_different_place_word2.append(j)
        
        return same_letters_same_place, [same_letters_different_place_word1, same_letters_different_place_word2]
```

```python
print("Let's compare some words!")
cont = True
while cont:
    word1, word2 = generate_two_words(words)
    matching_letters_same_place, matching_letters_diff_place = compare_words(word1, word2)
    print(f"Generated words: {word1}, {word2}")
    print(f"Same letters in the same place: {matching_letters_same_place}")
    print(f"Same letters in different places: {matching_letters_diff_place}")
    print("Enter to continue, 'q' to quit")
    if input().startswith('q'):
        print("Quit")
        cont = False
```

### Bài 3: Sudoku:

Cho 1 ô Sudoku bảng 9x9 chứa các số từ 1 đến 9. Nhiệm vụ là kiểm tra xem ô Sudoku bảng đã đúng hoàn toàn hay chưa bằng cách: Các hàng, cột và vùng 3x3 đều không có số nào trùng nhau.

```python
# Tạo 1 bảng Sudoku bảng 9x9 chứa các số từ 1 đến 9.
def generate_sudoku():
    # Start with an empty 9x9 grid
    grid = [[0 for _ in range(9)] for _ in range(9)]
    
    def is_valid(num, row, col):
        # Check if num can be placed at grid[row][col] without violations
        for i in range(9):
            if grid[row][i] == num or grid[i][col] == num:
                return False
        start_row, start_col = 3 * (row // 3), 3 * (col // 3)
        for i in range(3):
            for j in range(3):
                if grid[start_row + i][start_col + j] == num:
                    return False
        return True

    def fill_grid():
        # Fill the grid with numbers ensuring no violations in the initial state
        numbers = list(range(1, 10))
        for i in range(9):
            for j in range(9):
                random.shuffle(numbers)
                for num in numbers:
                    if is_valid(num, i, j):
                        grid[i][j] = num
                        break
                else:
                    return False
        return True
    
    # Generate a valid Sudoku grid first
    while not fill_grid():
        grid = [[0 for _ in range(9)] for _ in range(9)]
    
    # Introduce subtle violations
    def introduce_violations():
        violations = random.randint(1, 5)
        for _ in range(violations):
            row = random.randint(0, 8)
            col = random.randint(0, 8)
            num = random.randint(1, 9)
            grid[row][col] = num

    if random.choice([True, False]):
        introduce_violations()
    return grid

# In ra ô Sudoku
def print_grid(grid):
    for i, row in enumerate(grid):
        if i % 3 == 0 and i != 0:
            print("-" * 21)
        for j, num in enumerate(row):
            if j % 3 == 0 and j != 0:
                print("|", end=" ")
            print(num if num != 0 else ".", end=" ")
        print()

sudoku_grid = generate_sudoku()
print_grid(sudoku_grid)
```
