## Penalized Quadratic Partial Optimal Transport
**Quadratic Partial Optimal Transport (QPOT)**: thêm vào quadratic regularizer, bài toán được diễn giải ở dạng:
\begin{align}
\min_{\mathbf{X} \in U(\mathbf{r},\mathbf{c}, s)} &f_\eta(x) \triangleq \langle \mathbf{C}, \mathbf{X} \rangle  + \eta\|\mathbf{X}\|_2^2\\
\text{s.t:} \quad
&\sum_j \mathbf{X}_{i,j} \leq \mathbf{r} \\
&\sum_i \mathbf{X}_{i,j} \leq \mathbf{c} \\
&\sum_i \sum_j \mathbf{X}_{i,j} = s \\
&\mathbf{X} \geq 0
\end{align}
Gọi $U(\mathbf{r}, \mathbf{c}, s)$ là tập hợp các nghiệm $\mathbf{X}$ thỏa mãn 4 điều kiện trên: 

$$ U(\mathbf{r}, \mathbf{c}, s) = \{ \mathbf{X} \mid 1_n^T\mathbf{X}1_n = s, \, \mathbf{X}1_n \leq \mathbf{r}, \, \mathbf{X}^T1_n \leq \mathbf{c}, \, \mathbf{X} \geq 0\}$$

**Penalized Partial Optimal Transport (P-QPOT)**: Từ cách diễn giải trên của bài toán POT, chúng ta thử biến nó thành bài toán QPOT phạt 
\begin{align}
\min &F_{\eta, \alpha}(x) \triangleq f_\eta(x) + P(\mathbf{X}, \alpha)\\
\text{s.t:} \quad
&\sum_i \sum_j \mathbf{X}_{i,j} = s \\
&\mathbf{X} \geq 0
\end{align}
Với:
\begin{align*}
P(\mathbf{X}, \alpha) &= \alpha\left[max(0, \mathbf{X}1_n - r)^2 + max(0, \mathbf{X}^T1_n - c)^2\right] \\
                    &= \alpha \sum^n_{i=1}\left[min(0, r_i - (\mathbf{X}1_n)_i)^2 + min(0, c_i - (\mathbf{X}^T1_n)_i)^2\right]
\end{align*}

Gọi $F$ là tập hợp các $\mathbf{X}$ thỏa mãn:
$$F = \{\mathbf{X} \in \mathbb{R}^{n \times n}, \mathbf{X}1_n \leq \mathbf{r}, \mathbf{X}^T1_n \leq \mathbf{c}\}$$

### Bài tập 1: Viết dưới dạng hoàn chỉnh $P(\mathbf{X}, \alpha)$

Diễn giải đầy đủ thành dạng ma trận và vector các điều kiện $\mathbf{X}1_n \leq \mathbf{r}$ và $\mathbf{X}^T1_n \leq \mathbf{c}$, từ đó diễn giải đầy đủ công thức của $P(\mathbf{X}, \alpha)$:

Với $\mathbf{X} \in \mathcal{U}(\mathbf{r}, \mathbf{c}, s)= \{ \mathbf{X} \in \mathbb{R}_{+}^{n \times n}: \mathbf{X} 1_n \leq \mathbf{r}, \mathbf{X}^\top 1_n \leq \mathbf{c}, 1_n^\top \mathbf{X} 1_n = s \}$, ta viết lại 2 điều kiện ràng buộc như sau:

\begin{align*}
            \mathbf{X}\mathbf{1}_n \leq \mathbf{r}
            &\Leftrightarrow
            \mathbf{X} \mathbf{1}_n = \begin{bmatrix} 
            \sum_{j=1}^{n} \mathbf{X}_{1j} \\ 
            \sum_{j=1}^{n} \mathbf{X}_{2j} \\ 
            \vdots \\ 
            \sum_{j=1}^{n} \mathbf{X}_{nj} 
            \end{bmatrix} 
            \leq 
            \begin{bmatrix} 
            \mathbf{r}_1 \\ 
            \mathbf{r}_2 \\ 
            \vdots \\ 
            \mathbf{r}_n 
            \end{bmatrix} 
            = \mathbf{r}  \\
            &\Leftrightarrow \mathbf{r}_i - (\mathbf{X} \mathbf{1}_n)_i \geq 0, \quad \forall i
            \end{align*}

\begin{align*}
            \mathbf{X}^T \mathbf{1}_n \leq \mathbf{c} 
            &\Leftrightarrow 
            \mathbf{X}^T \mathbf{1}_n = \begin{bmatrix} 
            \sum_{i=1}^{n} \mathbf{X}_{i1} \\ 
            \sum_{i=1}^{n} \mathbf{X}_{i2} \\ 
            \vdots \\ 
            \sum_{i=1}^{n} \mathbf{X}_{in} 
            \end{bmatrix} 
            \leq 
            \begin{bmatrix} 
            \mathbf{c}_1 \\ 
            \mathbf{c}_2 \\ 
            \vdots \\ 
            \mathbf{c}_n 
            \end{bmatrix} 
            = \mathbf{c} \\
            &\Leftrightarrow \mathbf{c}_i - (\mathbf{X}^T \mathbf{1}_n)_i \geq 0, \quad \forall i
            \end{align*}


### Bổ đề 1: $P(\mathbf{X}, \alpha)$ là hàm đơn điệu với $\alpha$
Cho $0 < \alpha_1 < \alpha_2$, chúng ta sẽ có: 
$$ P(\mathbf{X}, \alpha_1) \leq P(\mathbf{X}, \alpha_2) $$

Ngoài ra, với $\alpha > 0$: 
\begin{align}
&P(\mathbf{X}, \alpha) = 0, \; \forall \mathbf{X} \in F \\
&P(\mathbf{X}, \alpha) > 0, \; \forall \mathbf{X} \notin F \\
lim_{\alpha \to \infty} &P(\mathbf{X}, \alpha) = \infty, \; \forall \mathbf{X} \notin F
\end{align}

**Proof**:  

Với $\mathbb{X} \in F = \{\mathbb{X}\in \mathbb{R}^{n \times n}: \mathbb{X} 1_n \leq \mathbb{r}, \mathbb{X}^\top 1_n \leq \mathbb{c}\}$, ta viết lại 2 điều kiện ràng buộc như sau:
\begin{align*}
\mathbf{X}\mathbf{1}_n \leq \mathbf{r}
&\Leftrightarrow \mathbf{r}_i - (\mathbf{X} \mathbf{1}_n)_i \geq 0, \quad \forall i \\
&\Leftrightarrow \min(\mathbf{r}_i - (\mathbf{X} \mathbf{1}_n)_i, 0) = 0 \quad \forall i \quad (1)
\\
\mathbf{X}^T \mathbf{1}_n \leq \mathbf{c} 
&\Leftrightarrow \mathbf{c}_i - (\mathbf{X}^T \mathbf{1}_n)_i \geq 0, \quad \forall i \\
&\Leftrightarrow \min(\mathbf{c}_i - (\mathbf{X}^T \mathbf{1}_n)_i, 0) = 0 \quad \forall i \quad (2)
\end{align*}
Từ (1), (2), ta có:
\begin{align*}
P(\mathbf{X}, \alpha)
&= \alpha \sum_{i=1}^{n} \left( \min \{ 0, \mathbf{r}_i - (\mathbf{X} \mathbf{1}_n)_i \}^2 + \min \{ 0, \mathbf{c}_i - (\mathbf{X}^T \mathbf{1}_n)_i \}^2 \right) \\
&= \alpha \sum_{i=1}^{n} \left( 0^2 + 0^2 \right) = 0
\end{align*}
Ngược lại, giả sử $\mathbb{X} \notin F$, khi đó, tồn tại $i \in \{1, \cdots, n\}$ sao cho một trong 2 trường hợp này xảy ra:
+ $\mathbf{r}_i - (\mathbf{X} \mathbf{1}_n)_i < 0$
+ $\mathbf{c}_i - (\mathbf{X}^T \mathbf{1}_n)_i < 0$.  

Khi đó: $$\min \{ 0, \mathbf{r}_i - (\mathbf{X} \mathbf{1}_n)_i \}^2 + \min \{ 0, \mathbf{c}_i - (\mathbf{X}^T \mathbf{1}_n)_i \}^2 = k > 0$$. 
Vì vậy, hàm phạt sẽ:, $$P(\mathbf{X}, \alpha)
= \alpha \sum_{i=1}^{n} \left( \min \{ 0, \mathbf{r}_i - (\mathbf{X} \mathbf{1}_n)_i \}^2 + \min \{ 0, \mathbf{c}_i - (\mathbf{X}^T \mathbf{1}_n)_i \}^2 \right) = \alpha c > 0$$

Cho $0< \alpha_1 < \alpha_2$, cộng với tính chất của hàm phạt:
+ $P(\mathbf{X}, \alpha) = 0$ for $\mathbb{X} \in F$
+ $P(\mathbf{X}, \alpha) > 0$ for $\mathbb{X} \notin F$

Thì điều này xảy ra
$$P(\mathbf{X}, \alpha_1) \leq P(\mathbf{X}, \alpha_2)$$
Với tính chất đồng biến theo $\alpha$ của hàm phạt khi $\mathbf{X} \notin F$ làm cho $P(\mathbb{X}, \alpha) = \alpha k,\; k > 0$, ta có: $$\lim_{\alpha\to\infty}P(\mathbb{X}, \alpha) =\lim_{\alpha\to\infty}\alpha k = \infty$$

### Định Lý 1: Mối quan hệ giữa bài toán QPOT và P-QPOT:
Ta có:
$$\textbf{QPOT}_{\eta}(\mathbf{r}, \mathbf{c}, s) =  \lim_{\alpha \to \infty} \textbf{P-QPOT}_{\eta, \alpha}(\mathbf{r}, \mathbf{c}, s)$$

**Proof**:  
Sử dụng tính chất
\begin{align*}
P(\mathbf{X}, \alpha) = 0, \forall \mathbf{X} \in F \Rightarrow & \lim_{\alpha \to \infty} P(\mathbf{X}, \alpha) = 0 , \forall \mathbf{X} \in F\\
\text{and:} \quad & \lim_{\alpha \to \infty} P(\mathbf{X}, \alpha) = \infty,  \forall \mathbf{X} \not\in F.
\end{align*}
Đặt $\mathbf{X} \in \mathcal{U}$ là nghiệm của bài toán $\textbf{P-QPOT}_{\eta, \alpha}(\mathbf{r}, \mathbf{c}, s)$, $\mathbf{X}$ lúc đó đồng thời sẽ thỏa mãn các điều kiện của tập $F$.  
Từ đó, $\lim_{\alpha \to \infty}\textbf{P-QPOT}_{\eta, \alpha}(\mathbf{r}, \mathbf{c}, s)$ có thể được viết lại thành:
\begin{align*}
    \lim_{\alpha \to \infty}\textbf{P-QPOT}_{\eta, \alpha}(\mathbf{r}, \mathbf{c}, s) &= \lim_{\alpha \to \infty} \left( \textbf{QPOT}_{\eta}(\mathbf{r}, \mathbf{c}, s) + P(\mathbf{X}, \alpha) \right) \\
    &= \textbf{QPOT}_{\eta}(\mathbf{r}, \mathbf{c}, s) + \lim_{\alpha \to \infty}P(\mathbf{X}, \alpha)\\
    &= \textbf{QPOT}_{\eta}(\mathbf{r}, \mathbf{c}, s) + 0\\
    &= \textbf{QPOT}_{\eta}(\mathbf{r}, \mathbf{c}, s)
\end{align*}

### Định Lý 2: $\varepsilon-convergence$ của bài toán P-QPOT với giá trị $\alpha$ đủ lớn

Ở Lecture trước, chúng ta đã nói về việc với bài toán tối ưu có hàm phạt, việc tìm một penalty coefficient $\alpha$ đủ lớn rất quan trọng trong việc tìm được nghiệm "gần tối ưu" cho bài toán.  
Để tìm một penalty coefficient như vậy, chúng ta có công thức:
$$ \alpha = \frac{\bar{\alpha}h}{\varepsilon}$$
Trong đó:  
+ $\bar{\alpha} > \frac{z - f_\eta(\bar{\mathbf{X}})}{\min_ig_i(\bar{\mathbf{X}})}$ với $g_i(\bar{\mathbf{X}})$ là các điều kiện ràng buộc và $\bar{\mathbf{X}}$ là một giá trị bất kì thỏa mãn các điều kiện ràng buộc
+ $z = 0$ là chặn trên của $-f_\eta(\mathbf{X}^0)$ với $\mathbf{X}^0$ là giá trị tối ưu của $-f_\eta(\mathbf{X})$
+ Từ Slatter's condition, ta biết $\min_i g_i(\bar{x}) \leq \zeta$
+ $h = \frac{\sqrt{2n} + 1}{2}$
+ $\varepsilon$ là giá trị $\varepsilon$ convergence
+ $f_\eta(\mathbf{X}) \leq s \|\mathbf{C}\|_\infty + \eta s^2$
+ $\zeta$ là tham số Zlatter của P-QPOT và $min_i g_i({\bar{X}})$ là điều kiện ràng buộc nhỏ nhất:  
$\zeta= \min\bigg\{ r_{min}, c_{min}, \frac{1}{n} \big( \min\{\|\mathbf{r}\|_1, \|\mathbf{c}\|_1\} -s \big) \bigg\} \\ \Rightarrow min_i g_i({\bar{X}}) \leq \zeta$

Vậy, để đạt $\varepsilon-convergence$, chúng ta đặt $\alpha > \frac{(z - f_\eta(\bar{\mathbf{X}}))h}{\min_ig_i(\bar{\mathbf{X}})(\varepsilon)} \geq \frac{(s\|\mathbf{C}\|_\infty + \eta s^2)(\sqrt{2n} + 1)}{2\zeta\epsilon}$ 

### Để chứng minh Định Lý 2, trước tiên ta chứng minh các bổ đề sau:

#### Bổ đề 2: Giới hạn của hàm $f_\eta(\mathbf{X})$
Với mọi $\mathbf{X} \in \mathcal{U}(\mathbf{r}, \mathbf{c}, s)$, chúng ta chứng minh được rằng $f_\eta(\mathbf{X})$ có chặn trên
$$f_\eta(\mathbf{X}) \leq s \|\mathbf{C}\|_\infty + \eta s^2$$
Với $\|\mathbf{C}\|_\infty = \max_{1 \leq i \leq n}\sum^n_{j=1}c_{ij}$ là giá trị của hàng có tổng lớn nhất của ma trận $\mathbf{C}$

**Proof:**

+ $$\|\mathbf{X}\|^2_2 = \sum^n_{i=1}\sum^n_{j=1}x_{ij}^2 \leq \sum^n_{i=1}\sum^n_{j=1}x_{ij}^2 + 2 \sum_{\substack{i=1 \\ (i, j) \neq (k, l)}}^{n} \sum_{j=1}^{n} x_{ij} x_{kl} \leq \left(\sum^n_{i=1}\sum^n_{j=1}x_{ij}\right)^2  \leq s^2$$
+ Vì $\|\mathbf{C}\|_\infty = \max_{1 \leq i \leq n}\sum^n_{j=1}c_{ij}\; \text{và}\; \mathbf{C} \in \mathbb{R}^{n \times n}_+$.  Nên $c_{ij} \leq \|\mathbf{C}\|_\infty \; \forall i, j$  
Từ đó:
$$\langle \mathbf{C}, \mathbf{X}\rangle = \sum^n_{i=1}\sum^n_{j=1}c_{ij}x_{ij} \leq  \sum^n_{i=1}\sum^n_{j=1}\|\mathbf{C}\|_\infty x_{ij} = \|\mathbf{C}\|\sum^n_{i=1}\sum^n_{j=1}x_{ij} = s\|\mathbf{C}\|_\infty \quad $$ 

#### Bổ đề 3: Rim condition of the transportation problem
Cho một bài toán OT bất kì với tập nghiệm $\Upsilon(\mathbf{a}, \mathbf{b}) = \{\mathbf{X} \in \mathbb{R}^{n \times n}_+: \mathbf{X}1_n = a, \mathbf{X}^T1_n = b\}$ (lưu ý rằng tập nghiệm không bao gồm điều kiện $1^T_n\mathbf{X}1_n = 1$)  
Tập nghiệm $\Upsilon(\mathbf{a}, \mathbf{b})$ là một tập nghiệm hợp lệ khi và chỉ khi $\|\mathbf{a}\|_1 = \|\mathbf{b}\|_1$ (Khối lượng vận chuyển đi bằng khối lượng vận chuyển đến)

#### Bổ đề 4: Chứng minh tập nghiệm $\mathcal{U}(\mathbf{r}, \mathbf{c}, s)$ là một tập nghiệm hợp lệ

Với tập nghiệm $\mathcal{U}(\mathbf{r}, \mathbf{c}, s) = \{ \mathbf{X} \mid 1_n^T\mathbf{X}1_n = s, \, \mathbf{X}1_n \leq \mathbf{r}, \, \mathbf{X}^T1_n \leq \mathbf{c}, \, \mathbf{X} \geq 0\}$ là tập nghiệm của $f_\eta(\mathbf{X})$,  
Cho bất kì $\mathbf{r}, \mathbf{c} \in \mathbb{R}^n_+$ và $s > 0$, tập nghiệm $\mathcal{U}(\mathbf{r}, \mathbf{c}, s)$ luôn hợp lệ khi và chỉ khi $s \leq \min\{\|\mathbf{r}\|, \|\mathbf{c}\|\}$ (tổng khối lượng vận chuyển không vượt quá khối lượng mà nguồn có thể cung cấp hoặc đích đến có thể chứa)

**Proof:**

+ $\exist\mathbf{X} \in \mathcal{U}(\mathbf{r}, \mathbf{c}, s)$, khi đó $s \leq \min{\|\mathbf{r}\|_1, \|\mathbf{c}\|_1}$
+ Cho $s \leq \min{\|\mathbf{r}\|_1, \|\mathbf{c}\|_1}$, lúc đó, ta có thể tạo ra 2 giá trị $\mathbf{r}'$, $\mathbf{c}'$ khiến cho tập $\Upsilon(\mathbf{r}', \mathbf{c}')$ hợp lệ, và đồng thời $\Upsilon(\mathbf{r}', \mathbf{c}') \subseteq \mathcal{U}(\mathbf{r}, \mathbf{c}, s)$ và điều đó cũng làm $\mathcal{U}(\mathbf{r}, \mathbf{c}, s)$ hợp lệ

Với $\mathbf{X} \in \mathcal{U}(\mathbf{r}, \mathbf{c}, s)$, ta có:
\begin{align*}
    \mathbf{X} \in \mathcal{U} 
    &\Leftrightarrow 
    \begin{cases}
        \mathbf{X} 1_n \leq \mathbf{r} \\
        \mathbf{X}^T 1_n \leq \mathbf{c} \\
        1^T_n\mathbf{X}1_n = s
    \end{cases}\\
    &\Leftrightarrow 
    \begin{cases}
        \sum^i_n\sum^j_n x_{ij} \leq \|\mathbf{r}\|_1 \\
        \sum^j_n\sum^i_n x_{ij} \leq \|\mathbf{c}\|_1 \\
        \sum^i_n\sum^j_n x_{ij} = s
    \end{cases}\\
    &\Rightarrow 
    \begin{cases}
        s \leq \|\mathbf{r}\|_1 \\
        s \leq \|\mathbf{c}\|_1
    \end{cases}
    \Leftrightarrow s \leq \min\{\|\mathbf{r}\|_1, \|\mathbf{c}\|_1\}
\end{align*}

Ngược lại, với $s\leq \min\{ \|\mathbf{r}\|_1, \|\mathbf{c}\|_1 \}$. Chúng ta có thể dựng lên $\mathbf{r}'$ và $\mathbf{c}'$ sao cho:
+ $\mathbf{c}'_i \leq \mathbf{c}_i \; \forall i = 1,..., n$
+ $\mathbf{r}'_i \leq \mathbf{r}_i \; \forall i = 1, ..., n$
+ $\|\mathbf{r}'\|_1 \leq \|\mathbf{r}\|_1 , \, \|\mathbf{c}'\|_1 \leq \|\mathbf{c}\|_1$
+ $\|\mathbf{r}'\|_1 = \|\mathbf{c}'\|_1 = s$

Ví dụ, để đảm bảo điều kiện $\|\mathbf{r}'\|_1 = \|\mathbf{c}'\|_1 = s$ và $\|\mathbf{r}'\|_1 \leq \|\mathbf{r}\|_1 , \, \|\mathbf{c}'\|_1 \leq \|\mathbf{c}\|_1$, chúng ta có thể xây dựng  $\mathbf{r}'$ và $\mathbf{c}'$ sao cho:
+ $\mathbf{r}'_i = s\mathbf{r}_i/\|\mathbf{r}\|_1 \; i=1,...,n$
+ $\mathbf{c}'_i = s\mathbf{c}_i/\|\mathbf{c}\|_1 \; i=1,...,n$
 

Vì việc xây dựng $\|\mathbf{r}'\| = \|\mathbf{c}'\|$ là có thể, và thêm điều kiện $1^T_n\mathbf{X}1 = s$, như vậy tồn tại một domain hợp lệ $\mathbf{X} \in \Upsilon(\mathbf{r}', \mathbf{c}') = \left\{ \mathbf{X} \in \mathbb{R}_{+}^{n \times n} : \mathbf{X} 1_n = \mathbf{r}', \mathbf{X}^\top 1_n = \mathbf{c}', 1^T_n\mathbf{X}1 = s \right\}$. 

Ngoài ra, vì $\mathbf{r}'_i \leq \mathbf{r}_i \;\text{and} \; \mathbf{c}'_i \leq \mathbf{c}_i \; \forall i = 1,..., n$, chứng tỏ $\mathbf{X} 1_n = \mathbf{r}' \leq \mathbf{r}, \mathbf{X}^\top 1_n = \mathbf{c}' \leq \mathbf{c}$ làm cho $\Upsilon(\mathbf{r}', \mathbf{c}') \subseteq \mathcal{U}(\mathbf{r}, \mathbf{c}, s)$ và đồng thời làm cho $\mathcal{U}(\mathbf{r}, \mathbf{c}, s)$ hợp lý. Vậy, $\mathbf{X} \in \mathcal{U}(\mathbf{r}, \mathbf{c}, s)$

#### Bổ đề 5: Slater's condition
Bài toán P-QPOT là một bài toán thỏa mãn Slater's condition:  
Giả sử $s < \min\{\|\mathbf{r}\|_1, \|\mathbf{c}\|_1\}$.  
Đặt $r_{min} = \min_{i} \{r_i\}$ và $c_{min} = \min_{i} \{c_i\}$ và:
$$
    \zeta= \min\bigg\{ r_{min}, c_{min}, \frac{1}{n} \big( \min\{\|\mathbf{r}\|_1, \|\mathbf{c}\|_1\} -s \big) \bigg\}.
$$
Khi đó, tồn tại $\bar{\mathbf{X}}\in \mathcal{U}(\mathbf{r}, \mathbf{c}, s)$ thỏa mãn:
\begin{align}
    \bar{\mathbf{X}} 1_n + \zeta 1_n \leq \mathbf{r} \\
    \bar{\mathbf{X}}^\top 1_n + \zeta 1_n \leq \mathbf{c}.
\end{align}

**Proof:**

Sử dụng tính chất 
$$
    \zeta= \min\bigg\{ r_{min}, c_{min}, \frac{1}{n} \big( \min\{\|\mathbf{r}\|_1, \|\mathbf{c}\|_1\} -s \big) \bigg\}.
$$

Ta có:
\begin{align*}
&\zeta \leq \frac{1}{n} \big( \min\{\|\mathbb{r}\|_1, \|\mathbb{c}\|_1\} -s \big) \\
\Leftrightarrow &s + n\zeta \leq \min\{\|\mathbb{r}\|_1, \|\mathbb{c}\|_1\} \\
\Leftrightarrow &s + \|\zeta 1_n\|_1 \leq \min\{\|\mathbb{r}\|_1, \|\mathbb{c}\|_1\} \\
\Leftrightarrow &s \leq \min\{\|\mathbb{r}\|_1 - \|\zeta 1_n\|_1, \|\mathbb{c}\|_1 - \|\zeta 1_n\|_1\} \\
\Leftrightarrow &s \leq \min\{\|\mathbb{r} - \zeta 1_n\|_1, \|\mathbb{c} - \zeta 1_n\|_1\} \quad \text{(Do: $\|\mathbf{r}\|_1 = \|(\mathbf{r}  - \zeta 1_n) +   \zeta 1_n\|_1 \leq  \|\mathbf{r}  - \zeta 1_n\|_1 +   \|\zeta 1_n\|_1 \implies  \|\mathbf{r}\|_1 -\|\zeta 1_n\|_1 \leq \|\mathbf{r}  - \zeta 1_n\|_1$)}
\end{align*}

Điều kiện này là điều kiện cần và đủ để xây dưng một domain hợp lệ $\mathcal{U}(\mathbf{r}- \zeta 1_n, \mathbf{c}- \zeta 1_n, s)$  

Kết hợp với Bổ đề 4, ta chứng minh được rằng tồn tại $\bar{\mathbf{X}} \in \mathcal{U}(\mathbf{r}- \zeta 1_n, \mathbf{c}- \zeta 1_n, s)$:
\begin{align*}
\bar{\mathbf{X}}1_n + \zeta 1_n &\leq \mathbf{r} \\
\bar{\mathbf{X}}^T1_n + \zeta 1_n &\leq \mathbf{c}
\end{align*}
